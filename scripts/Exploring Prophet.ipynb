```python
# https://facebook.github.io/prophet/docs/quick_start.html

import pandas as pd
from prophet import Prophet
import yfinance as yf
```

```python
data = yf.download('ETH-USD','2019-01-01','2022-07-25')
```

```python
data['timestamp']=data.index
```

```python
p_Test = data[['timestamp','Close']]
p_Test = p_Test.rename(columns={'timestamp': 'ds', 'Close' : 'y'})
```

```python
p_Test.dtypes
```

```python
m = Prophet()
m.fit(p_Test)
```

```python
future = m.make_future_dataframe(periods=365)
```

```python
forecast = m.predict(future)
```

```python
fig1 = m.plot(forecast)
```
